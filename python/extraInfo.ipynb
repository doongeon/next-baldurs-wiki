```python
from bs4 import BeautifulSoup
import json
import copy
```

```python
fs = open('data/baldursGateWeapons.html', 'r')
doc = fs.read()
fs.close()
```

```python
doc
```

```python
soup = BeautifulSoup(doc, 'html.parser')
```

```python
weapons = soup.find_all("div", "wiki-heading-content")
weapons = weapons[3:]
```

```python
def getExtraInfo(weapon):
  div_wikiParagraph = weapon.find_all("div", "wiki-paragraph")
  comment = ""
  howToGet = ""

  if(div_wikiParagraph is not list or len(div_wikiParagraph) < 1): {"comment": comment, "howToGet": howToGet}

  moreInfo = " ".join([token.strip() for token in div_wikiParagraph[-1].text.split("\n") if(token.strip() != "")]).replace("/", "")
  if("입수 방법 " in moreInfo):
    comment = moreInfo.split("입수 방법 ")[0]
    howToGet = moreInfo.split("입수 방법 ")[1]
  else:
    comment = moreInfo

  return {"comment": comment, "howToGet": howToGet}

  



  
```

```python
props = []

for i, weapon in enumerate(weapons):
  name_ko = ""

  div_wikiParagraph = weapon.find_all("div", "wiki-paragraph")
  if(len(div_wikiParagraph) == 0): continue

  weaponNameText = div_wikiParagraph[1].text
  
  name_ko = [token for token in weaponNameText.replace("\n", "").strip().split("  ") if(token != "")][0]

  extraInfo = getExtraInfo(weapon)

  props.append({"nameKo" : name_ko, "comment": extraInfo["comment"], "howToGet": extraInfo["howToGet"]})

props
```

```python
for prop in props:
  if(prop["comment"] == "" and prop["howToGet"] == ""):
    print(prop["nameKo"])
```

```python

fs = open('gameData_v03.json', 'r')
weapons = json.load(fs)
fs.close()
```

```python
newWeapons = []

for weapon in weapons:
  newWeapon = copy.deepcopy(weapon)
  newWeapon["name_ko"]

  newWeapon["comment"] = ""
  newWeapon["howToGet"] = ""
  
  for prop in props:
    if(prop["nameKo"] == newWeapon["name_ko"]):
      newWeapon["comment"] = prop["comment"]
      newWeapon["howToGet"] = prop["howToGet"]

  if(newWeapon["comment"] == "" and newWeapon["howToGet"] == ""): 
    print(newWeapon["name_ko"])

  newWeapons.append(newWeapon)

```

```python
fs = open('gameData_v04.json', 'w')
json.dump(newWeapons, fs, ensure_ascii=False)
fs.close()
```
